```python
from bilby.core import prior
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
prior = prior.PriorDict().from_json('./BBH_15_parameters_O4_prior.json')
```

```python
N = 300
injection_parameters = prior.sample(N)
```

```python
injection_parameters
```

```python
df = pd.DataFrame(injection_parameters)
df['mass_2_source'] = df['mass_ratio'] * df['mass_1_source']
print(df)
```

```python
GW_df_full = pd.read_csv("filtered_GW_sources.csv")

gamma = stats.gamma
GW_a, GW_loc, GW_scale = 1.9094, 0, 1319.36
GW_probs = gamma.pdf(GW_df_full['luminosity distance'], GW_a, GW_loc, GW_scale)
GW_probs = GW_probs / np.sum(GW_probs)

GW_df_indices = np.random.choice(len(GW_df_full), N, replace=False, p=GW_probs)
GW_df = GW_df_full.iloc[GW_df_indices]
print(GW_df)
```

```python
df['luminosity_distance'] = np.array(GW_df['luminosity distance'])
df['dec'] = np.deg2rad(np.array(GW_df['Dec_deg']))
df['ra'] = np.deg2rad(np.array(GW_df['RA_deg']))
df['event'] = [str(f"{i:04d}") for i in range(1, N+1)]
#df['event'] = df['event']
print(df)
```
